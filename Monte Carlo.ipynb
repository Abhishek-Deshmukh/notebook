```python
from numpy import sqrt, linspace
from numpy.random import uniform

from IPython.display import display_latex, display

from bokeh import plotting
from bokeh.io import show, output_notebook

output_notebook()
```

```python
# the equation points
xs = linspace(0, 2, 1000)
ys = sqrt(4 - xs**2)

# random points
SIZE = 10000
pts = uniform(0, 2, size = (SIZE, 2))

# pi
inside = pts[:,0]**2 + pts[:,1]**2 < 4
pi = sum(inside)*4 / SIZE
display_latex(f'$$ \pi = {pi} $$', raw=True)

# the plot
fig  = plotting.figure()
fig.line(xs, ys, color="blue")
fig.scatter(pts[inside][:,0], pts[inside][:,1])
fig.scatter(pts[~inside][:,0],pts[~inside][:,1], color="red")
show(fig)
```
