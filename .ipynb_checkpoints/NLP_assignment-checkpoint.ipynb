```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def stream(function, x_range=(0, 10), y_range=(-5, 5), density=1, x_scale=20, xlabel="t(s)", ylabel="", title=""):
    """makes and shows a stream plot for the given function with other beautification parameters"""
    u = np.ones(((x_range[1]-x_range[0])*10, (y_range[1]-y_range[0])*10)).T*x_scale
    v = function(np.repeat([np.arange(*y_range, 0.1)], 100, axis=0)).T
    plt.streamplot(*np.meshgrid(np.arange(*x_range, 0.1), np.arange(*y_range, 0.1)), u, v, density=density)
    plt.xlabel(xlabel)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.show()
```

## 2.2.1
equation: $\dot{x}=4x^2-16$

```python
def q2_2_1(x):
    return 4*np.square(x) - 16
x = np.linspace(-8, 8, 100)
y = q2_2_1(x)
plt.plot(x, y)
plt.plot([-8, 8], [0, 0])
plt.xlabel("$x$")
plt.ylabel("$\dot{x}$")
plt.title("(2.2.1) $\dot{x} = 4x^2-16$")
plt.show()
```

2 stable points
- $x=2$ => unstable fixed point
- $x=-2$ => stable fixed point

```python
stream(q2_2_1, xlabel="t", ylabel="x", title="(2.2.1) Stream plot for $\dot{x}=4x^2-16$")
```

## 2.2.3
equation $\dot{x} = x - x^3$

```python
def q2_2_3(x):
    return x - np.power(x, 3)
x = np.linspace(-3, 3, 100)
y = q2_2_3(x)
plt.plot(x, y)
plt.plot([-3, 3], [0, 0])
plt.xlabel("$x$")
plt.ylabel("$\dot{x}$")
plt.title("(2.2.3) $\dot{x}=x - x^3$")
plt.show()
```

fixed points:
- $x=-1$ => stable fixed point
- $x=0$ => unstable fixed point
- $x=1$ => stable fixed point

The value is generally low in the region between -1 and 1 so at large it will act as a stable fixed point on a large scale.

```python
stream(q2_2_3, y_range=(-2, 2), x_scale=1, xlabel="t", ylabel="x", title="(2.2.3) Stream plot for $\dot{x}=x-x^3$")
```

## 2.2.13
Analytical solution (TODO)

```python
def q2_2_13_c(x, k, m, g=9.8):
    return g - (k/m)*np.square(x)
x = np.linspace(-5, 5, 100)
y = q2_2_13_c(x, 1, 1)
plt.plot(x, y)
plt.plot([-5, 5], [0, 0])
plt.xlabel("$v$")
plt.ylabel("$\dot{v}$")
plt.title("(2.2.13) $\dot{v} = g-kv^2/m$ $(m=1, k=1, g=9.8)$")
plt.show()
```

Fixed points:
- $x = -\sqrt{\frac{mg}{k}}$ => unstable fixed point
- $x = \sqrt{\frac{mg}{k}}$ => stable fixed point

```python
stream(lambda x: q2_2_13_c(x, 1, 1), y_range=(-7, 7), x_scale=5, xlabel="t", ylabel="v", title="(2.2.13) Stream plot for $\dot{v}=g-kv^2/m$")
```

## 2.3.4

```python
def q2_3_4_b_n_dot(n, r, a, b):
    return r*n - a*np.square(n-b)*n
x = np.linspace(0, 1.5, 100)
y = q2_3_4_b_n_dot(x, 3, 30, 1)
plt.plot(x, y)
plt.plot([0, 1.5], [0, 0])
plt.xlabel("$N$")
plt.ylabel("$\dot{N}$")
plt.title("(2.3.4) $\dot{N}=rN - aN(N-b)^2$")
plt.show()
```

Fixed points:
- $n=0$ => stable fixed point
- $n=b-\sqrt{\frac{r}{a}}$ => unstable fixed point
- $n=b+\sqrt{\frac{r}{a}}$ => stable fixed point

```python
stream(lambda x: q2_3_4_b_n_dot(x, 3, 30, 1), y_range=(0, 2), x_scale=5, xlabel="t", ylabel="N", title="(2.3.4) Stream plot of $\dot{N}=rN - aN(N-b)^2$")
```

```python
def q2_3_4_b_eg(n, r, a, b):
    return r - a*np.square(n-b)
x = np.linspace(0, 2, 100)
y = q2_3_4_b_eg(x, 3, 30, 1)
plt.plot(x, y)
plt.plot([0, 2], [0, 0])
plt.xlabel("$N$")
plt.ylabel("$\dot{N}/N$")
plt.title("(2.3.4) $\dot{N}/N=r-a(N-b)^2$")
plt.show()
```

Fixed points:
- $n=b-\sqrt{\frac{r}{a}}$ => unstable fixed point
- $n=b+\sqrt{\frac{r}{a}}$ => stable fixed point

```python
stream(lambda x: q2_3_4_b_eg(x, 3, 30, 1), y_range=(-1, 3), x_scale=20, xlabel="t", ylabel="$\dot{N}/N$", title="(2.3.4) Stream plot for $\dot{N}/N=r-a(N-b)^2$")
```
